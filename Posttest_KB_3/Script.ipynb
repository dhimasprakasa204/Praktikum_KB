# DIAGRAM PIE VEHICLE YEAR

```python
data_pie = df['vehicle.year'].value_counts()
```

```python
data_pie = df['vehicle.year'].value_counts().reset_index(name = "year")
```

```python
plt.pie(data_pie.year,
    labels = data_pie["year"], 
    startangle=110, 
    autopct='%.1f%%', 
    shadow=True
)
plt.title('Diagram Pie Vehicle Year')
plt.show()
```

# korelasi Atribut Dengan Tipe Data Angka

```python
plt.figure(figsize=(10,5))

sns.heatmap(
    df.corr(),
    annot=True,
    cmap='flare'
) 
plt.title('korelasi antar attribut variabel ')

plt.show()
```

# Perbandingan Vehicle Year dan Renter TripsTaken

```python
plt.figure(figsize=(15,10))

x = df['vehicle.year']
y = df['renterTripsTaken']

plt.bar(x, y, color='green')

plt.title('Perbandingan Vehicle Year dan Renter Trips Taken', size=16)
plt.ylabel('Renter Trips Taken', size=14)
plt.xlabel('Vehicle Year', size=14)
plt.xticks(size=12)
plt.yticks(size=12)

plt.show()
```
